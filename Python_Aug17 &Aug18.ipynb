### functools(partial, lru_cache) and map, reduce and filter in python

```python
import functools
from functools import partial, lru_cache
```

```python
def f(a, b, c, x):
    return 1.5*a + 2*b + 10*c + x

g = partial(f, 21, 1)
g(2, 3)
```

```python
Lru - least recently used cache
```

```python
@lru_cache(maxsize=128)
def fib(n):
    if n<2:
        return n
    return fib(n-1) + fib(n-2)

fib(1000)
```

```python
fib(10)
```

```python
from functools import reduce
```

```python
n=5
reduce(lambda a, b: (a*b), range(1, n+1))
```

```python
l = [1,2,3,4,5, 6]
def add():
    sum = 0
    for i in l:
        sum = sum+i
    return sum
```

```python
import string
alphabet = string.ascii_letters + string.digits

while True:
    password = ''.join(choice(alphabet) for i in range(12))
    if (any(i.islower() for i in password)
    and any(c.isupper() for c in password)
    and sum(c.isdigit() for c in password) >=3):
        break

password
```

```python
from sqlite3 import Error
```

```python
def create_connection(db_file):
    """ create a database connection to a SQLite database"""
    connection = None
    try :
        connection = sqlite3.connect(db_file)
        print(sqlite3.version)
    except Error as e:
        print(e)
    finally:
        if connection:
            connection.close()
```

```python
create_connection("./example.db")
```

```python
def create_table(connection, create_table_sql):
    try:
        c = connection.cursor()
        c.execute(create_table_sql)
    except Error as e:
        print(e)
```

```python
def main():
    database = "./example.db"

    sql_create_projects_table = """ CREATE TABLE IF NOT EXISTS projects (
    id integer PRIMARY KEY,
    name text NOT NULL,
    begin_date text,
    end_date text
    );"""

    connection = sqlite3.connect(database)

    if connection is not None:
        create_table(connection, sql_create_projects_table)
    else:
        print("Error! Cannot create DB connection")

main()
```
